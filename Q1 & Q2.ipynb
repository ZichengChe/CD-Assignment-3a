```python
# Question 1: Which countries fall into the top and bottom categories based on internet usage in 2018? 
import pandas as pd
url = 'https://github.com/ZichengChe/CD-Assignment-a/raw/main/List%20of%20Countries%20by%20number%20of%20Internet%20Users%20-%20Sheet1.csv'
df = pd.read_csv(url)

df['Internet Users'] = pd.to_numeric(df['Internet Users'].str.replace(',', ''), errors='coerce')

top_countries = df.nlargest(10, 'Internet Users')
bottom_countries = df.nsmallest(10, 'Internet Users')


top_countries[['Country or Area', 'Internet Users']], bottom_countries[['Country or Area', 'Internet Users']]
```

```python

# Question 2: Is there a correlation between the "Population" and "Internet Users
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = 'https://github.com/ZichengChe/CD-Assignment-a/raw/main/List%20of%20Countries%20by%20number%20of%20Internet%20Users%20-%20Sheet1.csv'
df = pd.read_csv(url)

# Select the top 20 countries 
top_countries = df.sort_values(by='Internet Users', ascending=False).head(20)

# Create a Scatter Plot for the Top 20 Countries
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='Internet Users', data=top_countries, hue='Country or Area', palette='viridis')
plt.title('Scatter Plot: Top 20 Countries - Population vs Internet Users')
plt.xlabel('Population')
plt.ylabel('Internet Users')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
